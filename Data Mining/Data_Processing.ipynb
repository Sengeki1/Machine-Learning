```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('merged_data.csv', low_memory=False)
```

```python
df.tail()
```

```python
df.drop_duplicates(subset='balance', inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
df.replace(['', ' '], np.nan, inplace=True)  # Replace empty strings and spaces with NaN
df.dropna(inplace=True)  # Drop rows with NaN values

df.reset_index(drop=True, inplace=True)
filtered_df = df
filtered_df
```

```python
filtered_df.to_csv('filtered_data.csv', index=False)
```
